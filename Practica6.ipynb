1.- Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central



````python
#Bibliotecas de Python utilizadas para el análisis y visualización de datos.
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import statistics as stats
````

````python
autos=pd.read_csv("autos2.csv", encoding="ISO-8859-1")
````

````python
autos.head()
````

````python
autos.info()
````

````python
#Determinar el valor máximo de precios
max_precios=autos.price.max()
print(max_precios)
````

````python
#Determinar el valor mínimo de precios
min_precios=autos.price.min()
print(min_precios)
````

````python
#Determinar el valor de la moda de precios
moda=stats.mode(autos.price)
print(moda)
````

````python
#Determinar el valor de la media de precios
media=autos.price.mean()
print(media)
````

````python
#Determinar el valor de la mediana de precios
mediana=autos.price.median()
print(mediana)
````

````python
#Determinar el valor de la desviación estándar de precios
desviacion_estandar=autos.price.std()
print(desviacion_estandar)
````

````python
#Determinar los estadísticos y cuartiles con la función describe()
autos.price.describe()
````

2.-Realizar un Histograma de precios de los automóviles

````python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
import math
````

````python
df = pd.DataFrame(autos)
df
````

````python
#Regla de sturges
n = len(autos)
K = int(1 + math.log2(n))
print("Valor de n =", n, " \nNúmero de intervalos =", K)
````

````python
#Importo lib.
import matplotlib.pyplot as plt
from scipy import stats
````

````python
#Crear un histograma basado en la tabla de frecuencia

plt.hist(autos.price, bins=K, alpha=0.2,color="b", edgecolor="black")
plt.axvline(media, color="b", linestyle="dashed", linewidth=2, label=f"Media={media:.2f}")
plt.axvline(mediana, color="r", linestyle="dashed",linewidth=2, label=f"Mediana={mediana:.2f}")
plt.axvline(moda, color="y", linestyle="dashed",linewidth=2, label=f"Moda={moda:.2f}")
plt.axvline(desviacion_estandar, color="brown", linestyle="dashed",
            linewidth=2, label=f"Desviacion estandar={desviacion_estandar:.2f}")
plt.ylabel("Frecuencia")
plt.xlabel("Precios")
plt.title("Histograma de precios de los automóviles basado en la regla de Sturges")
plt.legend()
plt.show()
````



##` Interprete los datos y proporcione sus conclusiones`

En el histograma se dispersan los precios de 205 autos, se dividieron en 8 columnas (bins) usando la regla de Sturges proporcionada por el maestro.

Este histograma muestra que hay cerca de 100 autos
que tienen un precio en parte bajo al estar en el primer rango de precios.

En conclusión entre menor sea el precio más autos hay y entre mayor sea el precio menos son los autos hay con esos precios.

3.-Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)

````python
#Crear un diagrama de cajas para visualizar la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)
sns.boxplot(x="price", y="body-style", data=autos)
plt.title("Relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style) ")
plt.xlabel("Precio")
plt.ylabel("Tipo de vehículo")
````



##```Interpretación y conclusión```
Este diagrama de cajas y bigotes presenta cómo se reparten los precios de los autos dependiendo del tipo de vehículo.

Por ejemplo, los convertibles tienden a ser más costosos y la distribución en sus precios es variada, por otro lado los hardtops tienen una gran amplitud de dispersión en sus precios.

Por su parte los valores atípicos, que son los circulitos blancos que están del lado derecho , se muestran cuando algún auto es más caro que los demás.

Para concluir, con esta información ,puedo decir que el tipo de vehículo tiene mucho que ver en el precio, por lo visto los convertibles y hardtops son los más costosos.

4.-Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)

````python
#Crear un diagrama de cajas para visualizar la relación estadística entre el precio del auto (price) y tipo tracción ruedas (drive wheels)
sns.boxplot(x="price", y="drive-wheels", data=autos)
plt.title("Relación estadística entre el precio (price) y tipo tracción ruedas (drive wheels)")
plt.xlabel("Precio")
plt.ylabel("Tipo de tracción de ruedas")
````



##```Interpretación y conclusión```
En este diagrama de cajas y bigotes se puede observar que en el eje x están los precios y en el eje y el tipo de tracción en las ruedas, donde 4wd es tracción en las cuatro ruedas, fwd es tracción delantera y rwd es tracción trasera.

En rwd (tracción trasera) muestra el rango de precios más altos,y también que hay diversos valores atípicos que enseñan que algunos autos de tracción trasera son más caros que  otros valores atípicos.

La fwd tiene menos distribución en sus precios y se encuentra en precios un poco más baratos que 4wd.

Cada uno de los tipos de tracción de ruedas tiene sus propios valores atípicos el que menos cuenta con valores atípicos es 4wd y el mayor valor atípico es de rwd.

Por último el 4wd, es el que menos datos tiene,con la mínima cantidad de valores atípicos.

5.-Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y la ubicación del motor (engine-location)

Lo que nos enseña el diagrama de cajas y bigotes, es que es  más común el motor delantero ya que hay más cantidad de datos en comparación del motor trasero.

Para continuar, en el motor delantero hay más valores atípicos, que nos dicen que hay más motores ubicados en la parte de adelante con un precio alto de $30,000 (aprox.) en adelante.

Con la información obtenida del diagrama, se puede visualizar que los automóviles con motor ubicado en la parte de atrás son de manera destacada más costosos que los que tienen motor delantero.

Para terminar los autos con motor delantero tienen precios más variados que los autos con motor trasero; los carros que tienen el motor en la parte de atrás son más limitados.
