#**`PROBLEMA 1`**

>*La solución es de tipo Poisson*



La distribución de Poisson se utiliza para describir situaciones donde contamos eventos que ocurren en un periodo específico de tiempo o en un espacio definido.

Para poder utilizar esta distribución, es necesario que se cumplan ciertas condiciones:
- Se cuentan eventos que ocurren en un intervalo fijo de tiempo o espacio (como las personas que ingresan a diario).

*   Cada evento debe suceder de forma independiente unos de otros.

- La tasa promedio de ocurrencia, es decir, el número de eventos por día, debe ser constante. En este caso, hablamos de un promedio de 5 personas al día.
- Además, la probabilidad de que ocurran varios eventos al mismo tiempo es prácticamente cero o nula.

Y este problema cumple con todos los requisitos

---



````python
#Importo librerías
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
media = 5

# a) Probabilidad cuando x es igual a 2
prob_2 = poisson.pmf(2, media)
print("a) La probabilidad de que ingresen 2 personas es: " ,prob_2)

# b) Probabilidad cuando x excede los 10
prob_10 = poisson.sf(10, media)
print("b) La probabilidad de que ingresen más de 10 personas es: ",prob_10)
print(" ")

if prob_10 < 0.05:
    print("Es poco probable que ingresen más de 10 personas en un día.")
else:
    print("Es muy probable que ingresen más de 10 personas en un día.")
````



###Explicación de la pregunta b)
```
Si la probabilidad es muy baja ( menor a 0.05 o 5%),
se considera que el evento es muy poco probable que pase.
```





````python
# Parámetro de la distribución de Poisson
parametro = 5

# Rango de valores posibles de X (número de personas ingresadas)
# De 0 a 14 personas
x = np.arange(0, 15)
# Probabilidades para cada valor de X
probs = poisson.pmf(x, parametro)

# Código para hacer la gráfica
plt.bar(x, probs, color='pink', edgecolor='black')
plt.title('Distribución de Poisson (λ = 5)')
plt.xlabel('Número de personas ingresadas en un día')
plt.ylabel('Probabilidad')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.8)

#Hacer que los datos que buscamos sean notorios
plt.axvline(2, color='b', linestyle='--', label='x = 2')
plt.axvline(10, color='m', linestyle='--', label='x = 10')
plt.legend()
plt.show()
````

#**`PROBLEMA 2`**

>*La solución es de tipo Binomial*


La distribución binomial se usa cuando quieres saber la probabilidad de que ocurran k éxitos en n ensayos, bajo condiciones bien definidas y repetibles.

Para poder utilizar esta distribución, es necesario que se cumplan ciertas condiciones:


*   El número fijo de intentos (1000 motores).
* Cada intento (motor) tiene dos posibles resultados: si está defectuoso o no está defectuoso.

* La probabilidad de defecto es constante, es decir que no cambia:  
𝑝=0.001

* Los intentos son independientes entre ellos mismos.


Y este problema cumple con todos los requisitos

---



````python
import scipy.stats as stats

# Parámetros
n = 1000  # Número de motores
p = 0.001  # Probabilidad de que un motor sea defectuoso

# Probabilidad de que ninguno sea defectuoso (u = 0)
prob_0_defectuoso = stats.binom.pmf(0, n, p)

# Probabilidad de que 3 sean defectuosos (u = 3)
prob_3_defectuosos = stats.binom.pmf(3, n, p)

# Probabilidad de que 4 sean defectuosos (u = 4)
prob_4_defectuosos = stats.binom.pmf(4, n, p)

# Resultados
print("Probabilidad de que ninguno sea defectuoso: ",prob_0_defectuoso)
print("Probabilidad de que 3 sean defectuosos: ",prob_3_defectuosos)
print("Probabilidad de que 4 sean defectuosos: ",prob_4_defectuosos)
````
