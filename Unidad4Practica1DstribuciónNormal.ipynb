##**Ejercicio 1:**

Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.

 Hallar lo siguiente:

*   La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
*   Cuantos estudiantes pesan más de 68 y menos de 72 kg.





---



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Generar datos para la distribución normal
media= 70
desv_std=4 #Desviación estandar
num_datos=500

#generamos los datos de manera aleatoria y lo ponemos en una lista (array)
datos=np.random.normal(media,desv_std,num_datos)
datos[0:50]#Mostrar los 1eros 50
```

```python
#Crear un rango de valores para crear una densidad
x=np.linspace(media-3*desv_std,media+3*desv_std,num_datos)
#Calcular la pdf de la distribucion normal
pdf=1/(desv_std*np.sqrt(2*np.pi))*np.exp(-(x-media)**2/(2*desv_std**2))
```

```python
#Para no imprimir toda la lista de valores (500),desplegamos los primeros 50
print("Imprimir los 50 primeros valores de x", x[0:50], " , y los primeros 50 del pdf ", pdf[0:50])
```

```python
#Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15, alpha=0.4,color="b") #Histograma de los datos
plt.title("Distribución Normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show ()
```

```python
#Visualizar la PDF
plt.figure(figsize=(6,4))
plt.plot(x,pdf,label="Normal:(media:{};ds:{})".format(np.round(media,2),np.round(desv_std,2)))
plt.title("Función de Densidad de Probabilidad (PDF) - Distribución Normal")
plt.xlabel("Peso en Kg. de los estudiantes")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15, alpha=0.4,density=True, color="b")
plt.plot(x,pdf,"k",linewidth=2)

plt.title("Distribución Normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show()
```

##**La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.**

Una revision de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70 kilogramos y la desviacion
estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.

Hallar lo siguiente:



* La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.




---



##**La probabilidad de estudiantes que pesan menos de 72 kg y cuántos**

```python
plt.figure(figsize=(6,4))
#Parámetros de la distribución normal
media=70 #Media
desv_std=4 #Desviación Estándar
total_estudiantes= 500

#VARIABLE para Cálculo de la probabilidad de que un estudiante pese menos de 72 kg
peso_limite=72
#Primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 72 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_72kg=norm.cdf(peso_limite,media,desv_std)

#Cálculo del número de estudiantes que pesan menos de 72 kg
estudiantes_menos_72kg=total_estudiantes*prob_menos_72kg

print ("Probabilidad de que un estudiante pese menos de 72kg: ", prob_menos_72kg)
print ("Número de estudiantes que pesan menos de 72kg: ", estudiantes_menos_72kg)

#Gráficar la distribución normal
x=np.linspace(media-4*desv_std,media+4*desv_std,1000)
y=norm.pdf(x,media,desv_std)

plt.plot(x,y,"b",linewidth=2)
plt.fill_between(x,y,where=(x <=peso_limite),color="skyblue",alpha=0.5)
plt.title("Probabilidad  de que un estudiante pese menos de 72kg. \n $\mu=70$, $\sigma=4$, X=500")
plt.xlabel("Peso(kg)")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
plt.show()
```

##**La probabilidad de estudiantes que pesan más de 68kg y cuántos**

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 68 # Peso Límite
total_estudiantes = 500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 68 kg
prob_menos_68kg = norm. cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese mas de 68 kg es complementaria
prob_mas_68kg = 1 - prob_menos_68kg

# Calculamos el número aproximado de estudiantes que pesan más de 68 kg
estudiantes_mas_68kg = total_estudiantes * prob_mas_68kg

print("Probabilidad de que un estudiante pese más de 68 kg:", prob_mas_68kg)
print("Número aproximado de estudiantes que pesan más de 68 kg:", estudiantes_mas_68kg)

# Calcular la función de densidad de probabilidad de la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf (x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese más de 68 Kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad' )
plt.grid(True)
plt. show()
```

##**La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg y cuántos**

```python
# Límites del intervalo
lim_inf = 68
lim_sup = 72

# Calcular probabilidad P(68 < X < 72)
probabilidad = norm.cdf(lim_sup, media, desv_std) - norm.cdf(lim_inf, media, desv_std)

# Calcular número de estudiantes en ese rango
estudiantes_en_rango = round(probabilidad * num_datos)

# Mostrar resultados
print(f"Probabilidad de que un estudiante pese entre 68 y 72 kg: {probabilidad:.4f}")
print(f"Cantidad esperada de estudiantes en ese rango: {estudiantes_en_rango} estudiantes")

# Gráfica
x = np.linspace(media - 4*desv_std, media + 4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Distribución Normal', color='blue')
plt.fill_between(x, y, where=(x > lim_inf) & (x < lim_sup),color='skyblue', alpha=0.5,
                label='Entre 68 y 72 kg')
plt.title('Distribución de peso de estudiantes del ITD')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
